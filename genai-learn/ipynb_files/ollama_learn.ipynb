```python
import os
from langchain_groq import ChatGroq
from langchain_core.messages import HumanMessage, SystemMessage
from dotenv import load_dotenv
load_dotenv()

groq_api_key = os.getenv("GROQ_API_KEY")
llm = ChatGroq(api_key=groq_api_key, model="openai/gpt-oss-20b")

messages = [
  SystemMessage(content="Translate the following from English to German."),
  HumanMessage(content="Good Evening, how is the weather today?")
]
```

```python
# Outpur Parser
from langchain_core.output_parsers import StrOutputParser
parser = StrOutputParser()
parser.invoke(llm.invoke(messages))

chain = llm | parser
chain.invoke(messages)
```

```python
# Prompt Templates
from langchain_core.prompts import ChatPromptTemplate

genericTemplate = "Translate the following from {language}"
prompt = ChatPromptTemplate.from_messages(
  [("system", genericTemplate), ("human", "{text}")]
)
prompt.invoke({
  "language": "Hindi",
  "text": "Good Morning, how are you?"
})

chain = prompt | llm | parser
chain.invoke({
  "language": "English to Hindi",
  "text": "Good Morning, how are you?"
})
```
